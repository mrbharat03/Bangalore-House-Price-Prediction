```python
df.drop(['area_type', 'availability', 'society', 'balcony'], axis = 1, inplace = True)
```

```python
df['location'].fillna(df['location'].mode().iloc[0], inplace = True)
```

```python
df['size'] = df['size'].fillna(df['size'].mode().iloc[0])
```

```python
df['bath'] = df['bath'].fillna(df['bath'].mean())
df.info()
```

```python
df['bhk'] = df['size'].str.split().str.get(0)
df['bhk'] = df['bhk'].astype(int)
df.head()
```

```python
df = df[df['bhk'] <= 20]
df.head()
```

```python
df['total_sqft'].unique()
```

```python
def convertRange(x):
    temp = x.split('-')
    if len(temp) == 2:
        return (float(temp[0]) + float(temp[1]))/2
    else:
        try:
            return float(temp[0])
        except:
            return None
```

```python
df['total_sqft'] = df['total_sqft'].apply(convertRange)
df['total_sqft'] = df['total_sqft'].fillna(df['total_sqft'].mean())
df['total_sqft']
```

```python
df.info()
```

```python
df['price_per_ft'] = df['price'] * 1e5 / df['total_sqft']
df.head()
```

```python
df['location'].apply(lambda x: x.strip())
locations = df['location'].value_counts()
locations_count_less_10 = locations[locations <= 10]
df['location'] = df['location'].apply(lambda x: 'others' if x in locations_count_less_10 else x)
```

```python
df['location'].value_counts()
```

```python
df = df[((df['total_sqft']/df['bhk']) >= 300)]
df.describe()
```

```python
def remove_outlier_sqft(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf['price_per_sqft'])
        st = np.std(subdf['price_per_sqft'])
        gen_df = subdf[(subdf['price_per_sqft'] >= (m - st)) & (subdf['price_per_sqft'] <= (m + st))]
        df_out = pd.concat([df_out, gen_df])
    return df_out
df = remove_outlier_sqft(df)
df.describe()
```

```python
def remove_outlier_bhk(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean' : np.mean(bhk_df['price_per_sqft']),
                'std' : np.std(bhk_df['price_per_sqft']),
                'count' : bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df['price_per_sqft'] < (stats['mean'])].index.values)
    return df.drop(exclude_indices, axis = 'index')
df = remove_outlier_bhk(df)
```
